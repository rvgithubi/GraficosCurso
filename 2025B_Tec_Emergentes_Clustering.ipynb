```python
import numpy as np #Procesameinto numerico con listas/arreglos/matrices
import matplotlib.pyplot as plt #visualización

from sklearn.cluster import AgglomerativeClustering  #Otro metodo

from scipy.cluster.hierarchy import dendrogram, linkage


```

```python
# prompt: add label to each point to previous plot, consider order/index of point as label

import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
plt.scatter(X2[:,0], X2[:,1])
plt.title('Data Points with Labels')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)

# Add labels to points
for i, (x, y) in enumerate(X2):
    plt.text(x, y, str(i+1), fontsize=10, ha='right', va='bottom')

plt.show()
```

```python
agglomerative_clustering = AgglomerativeClustering(n_clusters=None, linkage='single', distance_threshold=0) # Single linkage clustering
# Fit and predict the clusters
agglomerative_clustering.fit(X2)
```

```python
# Plotting the dendrogram
linked = linkage(X2, 'single')
labelList = range(1, len(X2) + 1)
plt.figure(figsize=(14, 7))
dendrogram(linked,
            orientation='top',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True)
plt.title('Dendrogram for Single Linkage Clustering')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()

# Print cluster labels for each data point
print("Cluster labels:", agglomerative_clustering.labels_)
```

```python
# prompt: could you to previous plot change to more single plot, show the original points, but coloring now by cluster stored in linked

import matplotlib.pyplot as plt
# Number of clusters you want to visualize (e.g., based on dendrogram)
# You can change this value to see different clusterings
n_clusters_to_show = 5

# Perform Agglomerative Clustering with a specific number of clusters
single_clustering_fixed = AgglomerativeClustering(n_clusters=n_clusters_to_show, linkage='single')
single_clustering_fixed.fit(X2)
cluster_labels = single_clustering_fixed.labels_

plt.figure(figsize=(10, 7))
plt.scatter(X2[:,0], X2[:,1], c=cluster_labels, cmap='coolwarm') #bwr, berlin
plt.title(f'Data Points Clustered with {n_clusters_to_show} Clusters (Single Linkage)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)

# Add labels to points
#for i, (x, y) in enumerate(X2):
#    plt.text(x, y, str(i+1), fontsize=12, ha='right', va='bottom')

plt.colorbar(label='Cluster Label')
plt.grid(False)
plt.show()
```

#Definir K en K-Means

Coeficiente Silueta (Metrica/Distribucion) --> Para obtener mejor K (Numero de cluster/grupo)

```python
# prompt: Usa el algoritmo x-means o equivalente para detectar K optimo del algoritmo K-means aplicado a la variable X

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np


max_k = 10

silhouette_scores = []  #Lista de silhouette scores

for k in range(2, max_k + 1):  # 2,10
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10) # Set n_init explicitly
        kmeans.fit(X2) #Entrenar

        score = silhouette_score(X2, kmeans.labels_)  #Calcula el Coeficiente
        silhouette_scores.append(score)




```

```python
silhouette_scores
```

```python
k_range=np.array([2,3,4,5,6,7,8,9,10])
optimal_k = k_range[np.argmax(silhouette_scores)]
```

```python
silhouette_scores
```

```python



print(f"Optimal K suggested by Silhouette Score: {optimal_k}")

    # You can optionally visualize the silhouette scores
plt.figure(figsize=(8, 5))

plt.plot(k_range, silhouette_scores, marker='o')  #2,3...10 vs Coef. Silueta

plt.title('Silhouette Score vs. Number of Clusters (K)')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.xticks(k_range)
plt.grid(True)
plt.show()
```
