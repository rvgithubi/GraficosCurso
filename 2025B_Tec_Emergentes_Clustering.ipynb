```python
import numpy as np #Procesameinto numerico con listas/arreglos/matrices
import matplotlib.pyplot as plt #visualización

from sklearn.cluster import AgglomerativeClustering  #Otro metodo

from scipy.cluster.hierarchy import dendrogram, linkage


```

```python
# Sample data
X = np.array([[1, 2], [1.5, 1.8], [5, 8], [8, 8], [1, 0.6], [9, 11],[6,2],[7,2]])
```

```python
valores=[]  #Lista Python
for k in range(40):  #0,1,2,...9
   i=np.random.uniform(1,15)  #Aleatorio entre 1 y 15
   j=np.random.uniform(1,15)  #Aleatorio entre 1 y 15
   valores.append([i,j])

X=np.array(valores)  # Convertir Lista X en Array numpy
```

```python
# prompt: add label to each point to previous plot, consider order/index of point as label

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 7))
plt.scatter(X[:,0], X[:,1])
plt.title('Data Points with Labels')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)

# Add labels to points
for i, (x, y) in enumerate(X):
    plt.text(x, y, str(i+1), fontsize=12, ha='right', va='bottom')

plt.show()
```

```python
# Single linkage clustering
single_clustering = AgglomerativeClustering(n_clusters=None, linkage='single', distance_threshold=0)

# Fit and predict the clusters
single_clustering.fit(X)
```

```python
# Plotting the dendrogram
linked = linkage(X, 'single')
labelList = range(1, len(X) + 1)
plt.figure(figsize=(14, 7))
dendrogram(linked,
            orientation='top',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True)
plt.title('Dendrogram for Single Linkage Clustering')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()

# Print cluster labels for each data point
print("Cluster labels:", single_clustering.labels_)
```

```python
# prompt: could you to previous plot change to more single plot, show the original points, but coloring now by cluster stored in linked

import matplotlib.pyplot as plt
# Number of clusters you want to visualize (e.g., based on dendrogram)
# You can change this value to see different clusterings
n_clusters_to_show = 3

# Perform Agglomerative Clustering with a specific number of clusters
single_clustering_fixed = AgglomerativeClustering(n_clusters=n_clusters_to_show, linkage='single')
single_clustering_fixed.fit(X)
cluster_labels = single_clustering_fixed.labels_

plt.figure(figsize=(10, 7))
plt.scatter(X[:,0], X[:,1], c=cluster_labels, cmap='berlin') #bwr
plt.title(f'Data Points Clustered with {n_clusters_to_show} Clusters (Single Linkage)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)

# Add labels to points
for i, (x, y) in enumerate(X):
    plt.text(x, y, str(i+1), fontsize=12, ha='right', va='bottom')

plt.colorbar(label='Cluster Label')
plt.grid(False)
plt.show()
```
